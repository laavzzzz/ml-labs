```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
titanic = sns.load_dataset('titanic')

print("Titanic Dataset Sample:")
print(titanic.head())
```

```python
titanic_df = titanic[['survived', 'pclass', 'sex', 'age', 'fare', 'embarked']]

print("\nSelected Titanic Columns:")
print(titanic_df.head())
```

```python
# Cell 4 (FIXED): Handling missing values safely using .loc

# Numerical columns - mean
imputer_num = SimpleImputer(strategy='mean')
titanic_df.loc[:, ['age', 'fare']] = imputer_num.fit_transform(
    titanic_df.loc[:, ['age', 'fare']]
)

# Categorical columns - most frequent
imputer_cat = SimpleImputer(strategy='most_frequent')
titanic_df.loc[:, ['sex', 'embarked']] = imputer_cat.fit_transform(
    titanic_df.loc[:, ['sex', 'embarked']]
)

print("\nAfter handling missing values (Titanic) - WARNING FIXED:")
print(titanic_df.head())
```

```python
# Cell 5 (FIXED): Label Encoding using .loc to avoid SettingWithCopyWarning

label_encoder = LabelEncoder()

for col in ['sex', 'embarked']:
    titanic_df.loc[:, col] = label_encoder.fit_transform(
        titanic_df.loc[:, col]
    )

print("\nAfter Label Encoding (Titanic) - WARNING FIXED:")
print(titanic_df.head())
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title("Titanic Survival by Gender")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.scatterplot(data=titanic_df, x='age', y='fare', hue='survived')
plt.title("Titanic: Age vs Fare (Colored by Survival)")
plt.show()
```

```python
# Feature Scaling using StandardScaler (WARNING-FREE)

scaler = StandardScaler()

titanic_df.loc[:, ['age', 'fare']] = scaler.fit_transform(
    titanic_df.loc[:, ['age', 'fare']]
)

print("\nAfter Feature Scaling (Titanic) - WARNING FIXED:")
print(titanic_df.head())
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x='survived', y='age', data=titanic_df)
plt.title('Age vs Survival')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x='survived', y='fare', data=titanic_df)
plt.title('Fare vs Survival')
plt.show()
```

```python
plt.figure(figsize=(6, 5))
sns.boxplot(x='pclass', y='fare', data=titanic_df)
plt.title('Fare vs Passenger Class')
plt.show()
```
