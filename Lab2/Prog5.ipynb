```python
import numpy as np

data = [78, 82, 85, 90, 92, 94, 95, 100, 110, 120, 150]
data_sorted = sorted(data)

Q1 = np.percentile(data_sorted, 25)
Q3 = np.percentile(data_sorted, 75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = [x for x in data_sorted if x < lower or x > upper]

print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)
print("Lower Fence:", lower)
print("Upper Fence:", upper)
print("Outliers:", outliers)
```
