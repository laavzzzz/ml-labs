```python
height_ranges = [
    (59.95,61.95),
    (61.95,63.95),
    (63.95,65.95),
    (65.95,67.95),
    (67.95,69.95),
    (69.95,71.95),
    (71.95,73.95),
    (73.95,75.95)
]

frequencies = [5,3,15,40,17,12,7,1]

total_players = 100
```

```python
freq_less_than_65_95 = 0
for (low, high), freq in zip(height_ranges, frequencies):
    if high < 65.95:
        freq_less_than_65_95 += freq

percent_less_than_65_95 = (freq_less_than_65_95 / total_players) * 100 

print("percentage less than 65.95 inches:",percent_less_than_65_95)
```

```python
freq_between_61_95_and_65_95 = 0
for (low, high), freq in zip(height_ranges, frequencies):
    if low >= 61.95 and high <= 65.95:
        freq_between_61_95_and_65_95 += freq

percent_between = (freq_between_61_95_and_65_95 / total_players) * 100

print("Percentage between 61.95 and 65.95 inches:", percent_between)
```

```python
freq_between_61_95_and_71_95 = 0
for (low, high), freq in zip(height_ranges, frequencies):
    if low >= 61.95 and high <= 71.95:
        freq_between_61_95_and_71_95 += freq

print("Number of players between 61.95 and 71.95 inches:", freq_between_61_95_and_71_95)
```
