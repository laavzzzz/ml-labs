```python
# Pass = 1, Fail = 0
amit = [0, 0, 1, 0, 0, 1]
binit = [0, 1, 1, 1, 0, 1]

intersection = 0
union = 0

for a, b in zip(amit, binit):
    if a == 1 or b == 1:
        union += 1
    if a == 1 and b == 1:
        intersection += 1

jaccard = intersection / union
print("Jaccard Coefficient:", jaccard)
```
