```python
import numpy as np

data = np.array([5,10,11,13,15,35,50,55,72,92,204,215])

# Equal Frequency
equal_freq = np.array_split(data, 3)
print("Equal Frequency Bins:", equal_freq)

# Equal Width
min_val, max_val = data.min(), data.max()
width = (max_val - min_val) / 3

bins = [
    data[(data >= min_val) & (data < min_val + width)],
    data[(data >= min_val + width) & (data < min_val + 2*width)],
    data[(data >= min_val + 2*width)]
]

print("Equal Width Bins:", bins)
```
