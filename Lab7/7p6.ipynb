```python
import math
from collections import Counter

# Dataset: (Weight, Height, Species)
data = [
    (4, 35, "Cat"),
    (6, 40, "Dog"),
    (3, 25, "Cat"),
    (7, 45, "Dog"),
    (5, 30, "Cat"),
    (8, 50, "Dog"),
    (2, 20, "Cat"),
    (5, 35, "Dog")
]

# New animal to classify
new_point = (4, 30)
K = 5

# Step 1: Calculate Euclidean distances
distances = []
for weight, height, species in data:
    dist = math.sqrt((weight - new_point[0])**2 + (height - new_point[1])**2)
    distances.append((dist, species))

# Step 2: Sort by distance
distances.sort(key=lambda x: x[0])

# Step 3: Take K nearest neighbors
k_nearest = distances[:K]

# Step 4: Majority voting
species_count = Counter([species for _, species in k_nearest])
prediction = species_count.most_common(1)[0][0]

print("K nearest neighbors:", k_nearest)
print("Predicted Species:", prediction)
```
